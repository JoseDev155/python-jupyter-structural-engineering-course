```python
height = [174.3, 180.3, 183.4, 173.2, 186.4, 188.7, 175.8, 178.2, 177.2, 189.1]
weight = [65.7, 71.8, 76.4, 69.7, 78.5, 80.9, 71.2, 73.1, 74.9, 84.3]

plt.figure()
plt.scatter(height,
            weight,
            c = "b",
            ec = "k",
            alpha = 0.8,
            marker = "o",
            s = 200,
            lw = 0.5)
plt.title("Altura vs pesos")
plt.xlabel("Alturas $(cm)$")
plt.ylabel("Pesos $(Kg)$")
plt.show()
```

#### Line Plot

Para hacer un gráfico de líneas line plot, usamos el método `.plot()` del módulo `plt`.

Algunos de los parámetros de este método son:

* `x`: scalar, array o lista que indica la primera coordenada de las observaciones.
* `y`: Escalar, array o lista que indica la segunda coordenada de las observaciones
* `color`: para cambiar el color de relleno. También podemos referirnos a este parámetro por su diminutivo `c`
* `fmt`: para establecer un formato básico rápidamente como string. Por ejemplo, `or` son círculos rojos. El orden orden recomendado para introducir el formato de este modo es es `"[marker][linestyle][color]"` aunque también se admite `"[color][marker][linestyle]"`
* `linewidth`: para cambiar el grosor de la línea. También podemos referirnos a este parámetro por su diminutivo `lw`
* `linestyle`: para cambiar el estilo de la línea. También podemos referirnos a este parámetro por su diminutivo `ls`
* `alpha`: para cambiar la transparencia
* `marker`: para cambiar la **forma del punto**. Si no indicamos este parámetro, no se dibujan los puntos.
* `markersize`: para cambiar el tamaño de los puntos. También podemos referirnos a este parámetro por su diminutivo `ms`
* `markeredgecolor`: para cambiar el color del contorno del punto. También podemos referirnos a este parámetro por su diminutivo `mec`
* `markerfacecolor`: para cambia el color de relleno del punto. También podemos referirnos a este parámetro por su diminutivo `mfc`

Para el parámetro `fmt`, las opciones disponibles para `marker`, `linestyle` y `color` son

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 6, 4, 8, 10, 9, 7, 3, 5]

plt.figure(figsize=(10, 3.5))
plt.plot(x,
        y,
        c = "g",
        ls = ":",
        lw = 2,
        marker="o",
        ms = 10,
        mfc = "g")
plt.title("Alturas vs. pesos")
plt.xlabel("Alturas (cm)")
plt.ylabel("Pesos (Kg)")
plt.grid(True, lw = 0.5, ls = "--", c = "b")
plt.show()
```

Podríamos obtener el mismo resultado usando el parámetro `fmt`

```python
plt.figure()
plt.plot(x,
         y,
         "s--b",
         lw = 2,
         ms = 6, mfc = "m")
plt.title("Alturas vs. pesos")
plt.xlabel("Alturas (cm)")
plt.ylabel("Pesos (Kg)")
plt.show()
```

**Observación**. Si no indicamos parámetro `x`, se consideran como primeras coordenadas los números enteros 0, 1, 2, ..., n - 1, siendo n el número total
de observaciones:

```python
rainy_days = [20, 15, 19, 25, 13, 8, 3, 5, 10, 16, 20, 18]
plt.figure()
plt.plot(rainy_days, c = "b", marker = "v")
plt.title("Días de lluvia por mes", fontsize = 20, color = "g", fontweight = "bold")
plt.xlabel("Mes")
plt.ylabel("Total de días de lluvia")
plt.grid()
plt.show()
```

**Observación**. En un mismo plot podemos dibujar más de una línea:

```python
import numpy as np

l1 = np.array([3, 5, 1, 9])
l2 = np.array([8, 2, 6, 4])
plt.figure(figsize=(6, 7), dpi = 100, facecolor = "lightgray")
plt.plot(l1, c = "b", marker = "s", ms = 20, mfc = "g")
plt.plot(l2, c = "m", marker = "o", ms = 5, mfc = "k")
plt.title("2 líneas en 1 plot", fontsize = 15, color = "b", fontweight = "bold")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid()
plt.show()
```

#### Múltiples Plots
##### El método `.subplot()` de `plt`

Con el método `.subplot()` podemos dibujar múltiples plots en una misma figura.

Este método toma 3 parámetros que describen la layout de la figura. La layout se organiza en filas y columnas, que son representadas por los dos primeros
parámetros. El tercer parámetro representa el índice que ocupa el plot actual.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

# plot 1
l1 = np.array([3, 5, 1, 9])
plt.subplot(1, 2, 1)
plt.plot(l1, c = "m", ls = ":", lw = 2)

# plot 2
l2 = np.array([8, 2, 6, 4])
plt.subplot(1, 2, 2)
plt.plot(l2, c = "b", ls = "--", lw = 2)
plt.show()
```

**Observación**. En el ejemplo anterior, `plt.subplot(1, 2, 1)` indica que la layout tiene una fila, 2 columnas y que el plot que se dibujará a continuación
ocupa el primer plot. Mientras que `plt.subplot(1, 2, 2)`, indica de nuevo que la layout consta de una fila y dos columnas, pero en este caso el plot
subsiguiente se dibujará en el segundo plot.

Si por el contrario queremos una layout de la figura que consta de 2 filas y una columna, haríamos lo siguiente:

```python
# plot 1
plt.subplot(2, 1, 1)
plt.plot(l1, c = "m", ls = ":", lw = 2)

# plot 2
plt.subplot(2, 1, 2)
plt.plot(l2, c = "b", ls = "--", lw = 2)

plt.show()
```

##### Ejemplo 1

Vamos a hacer una layout que conste de 2 filas y 4 columnas. En total tendremos 8 subplots, de modo que los índices irán de 1 hasta 8:

```python
fig, (ax1, ax2) = plt.subplots(2, 4, figsize=(12, 8))

# plot 1
plt.subplot(2, 4, 1)
plt.plot(l1, c = "m", ls = ":", lw = 2)

# plot 2
plt.subplot(2, 4, 6)
plt.plot(l2, c = "b", ls = "--", lw = 2)

plt.show()
```

```python
# plot 1 (primera fila primera columna)
plt.subplot(2, 4, 1)
plt.plot(l1, c = "blue", ls = "-", lw = 2)

# plot 2 (primera fila segunda columna)
plt.subplot(2, 4, 2)
plt.plot(l2, c = "magenta", ls = "--", lw = 2)
plt.plot(l1, c = "blue", ls = "-", lw = 2)
plt.plot(l2, c = "cyan", ls = "-.", lw = 2)

# plot 3 (primera fila tercera columna)
plt.subplot(2, 4, 3)
plt.plot(l1, c = "black", ls = ":", lw = 2)

# plot 4 (primera fila cuarta columna)
plt.subplot(2, 4, 4)
plt.plot(l2, c = "cyan", ls = "-.", lw = 2)


# plot 5 (segunda fila primera columna)
plt.subplot(2, 4, 5)
plt.plot(l2, c = "black", ls = "--", lw = 2)

# plot 6 (segunda fila segunda columna)
plt.subplot(2, 4, 6)
plt.plot(l1, c = "cyan", ls = "-.", lw = 2)

# plot 7 (segunda fila tercera columna)
plt.subplot(2, 4, 7)
plt.plot(l2, c = "blue", ls = "-", lw = 2)

# plot 8 (segunda fila cuarta columna)
plt.subplot(2, 4, 8)
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

plt.show()
```

#### Añadiendo títulos

Podemos añadir un título a cada subplot con el método `.title()`

```python
# plot 1
l1 = np.array([3, 5, 1, 9])

plt.subplot(1, 2, 1)
plt.title("Subplot 1")
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

# plot 2
l2 = np.array([8, 2, 6, 4])

plt.subplot(1, 2, 2)
plt.title("Subplot 2")
plt.plot(l2, c = "blue", ls = "--", lw = 2)

plt.show()
```

Podemos añadir un título a la figura con el método `.suptitle()`

```python
# plot 1
plt.subplot(1, 2, 1)
plt.title("Subplot 1")
plt.xlabel("Eje X1")
plt.ylabel("Eje Y1")
plt.plot(l1, c = "magenta", ls = ":", lw = 2)

# plot 2
plt.subplot(1, 2, 2)
plt.title("Subplot 2")
plt.xlabel("Eje X2")
plt.ylabel("Eje Y2")
plt.plot(l2, c = "blue", ls = "--", lw = 2)

plt.suptitle("2 Subplots en una figura")

plt.show()
```

#### El método `.add_subplot()` de `plt`

Con el método `.add_subplot()` también podemos dibujar múltiples plots en una misma figura.

Para ello tendremos que seguir los siguientes pasos:

* Crear el objeto `figure` con el método `.figure()`
* Indicar el número de filas y de columnas que queremos con el método `.add_gridspec()`
* Crear tantos subplots como queramos mostrar con el método `.add_subplot()` e indicar las posiciones que van a tomar cada uno de ellos

```python
fig, axes = plt.subplots(2, 3)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(8,5))
# fig.subplots_adjust(hspace=0.8) # añade más espacio vertical entre suplots
fig.subplots_adjust(hspace=0.4, wspace=0.8) # añade más espacio vertical y horizontal entre suplots
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
x
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.figure(figsize=(10, 5))
plt.plot(x, y_sin, c = "b", ls = "--", lw = 3)
plt.plot(x, y_cos, c = "r", ls = ":", lw = 3)
plt.xlabel("x axis label")
plt.ylabel("y axis label")
plt.title("Seno and Coseno", fontsize = 15, color = "b", fontweight = "bold")
plt.legend(["$Sen(x)$", "$Cos(x)$"], loc = "lower left")
plt.grid(True, alpha = 0.7, linestyle = "-", color = "silver", linewidth = 0.8)
plt.savefig("sen,cos.pdf", dpi = 400, bbox_inches = "tight")
plt.show()
```
