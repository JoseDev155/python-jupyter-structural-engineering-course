```python
import matplotlib.pyplot as plt
import numpy as np
```

Factor de amplificacion dinamica en funcion del factor de amortiguamiento

```python
zeda = [0.01,0.03,0.05,0.1,0.2,0.7]
r = np.linspace(0,2.5,250)
plt.figure(figsize=(10,7))
plt.axvline(1, c = "r", ls = "--", lw = 1)
for zedai in zeda:
    D = 1/((1-r**2)**2+(2*zedai*r)**2)**0.5
    plt.plot(r,D,label=r'$\xi$ = '+str(zedai),lw=2)
plt.title('Factor de amplificacion dinamica en funcion del factor de amortiguamiento',pad = 12)
plt.grid(True, which="both", ls="-")
plt.minorticks_on()
plt.grid(which='minor', linestyle='-', color='gray', linewidth=0.2, alpha=0.6)
plt.grid(axis='x', linestyle='-', color='gray', linewidth=0.5, alpha=0.7)
plt.grid(axis='y', linestyle='-', color='gray', linewidth=0.5, alpha=0.7)
plt.ylabel('Factor de amplificacion dinamica,$D$')
plt.xlabel('Relaion de frecuencias, $r= w / w_n$')
plt.yscale('log')
plt.xlim(0,2.5)
plt.legend(fontsize=9,loc='upper right',frameon=True,facecolor='white',framealpha=1)
plt.ylim(0.1,100)
plt.text(1.6, 4.5, r'$D = \frac{1}{\sqrt{(1-r^2)^2+(2\xi r)^2}}$', fontsize=18)
plt.savefig('Factor_de_amplificacion_dinamica.png',dpi=500,bbox_inches='tight')
plt.show()
```
