```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Leer el archivo de excel

```python
df = pd.read_excel('Data/Story_Drifts.xlsx')
df
```

## Filtrar

Filtrar OutputCase **DERIVA X**

Filtrar StepType **Max**

Filtrar Direction **X**

```python
df1 = df[df.OutputCase.isin(["DERIVA X"])]
df1
```

```python
df1 = df1[df1.OutputCase.isin(["DERIVA X"])]
df1
```

```python
df1 = df1[df1.StepType.isin(["Max"])]
df1
```

```python
df1 = df1[df1.Direction.isin(["X"])]
df1
```

```python
df1 = df1.reset_index(drop=True)
df1
```

## Eliminar columnas

```python
eliminar_col = ["CaseType","StepType","StepNumber","Label","X","Y","Z"]
df1 = df1.drop(eliminar_col, axis=1)
df1
```

```python
new_row = {'Story': 'BASE','OutputCase': 'DERIVA X', 'Direction': 'X', 'Drift': 0.000}
df1 = df1._append(new_row, ignore_index=True)
df1
```

```python
limit = 0.007
df1["Limit"] = limit
df1["Drift"] = df1["Drift"].astype(float)
df1
```

### Trabajar con los resultados

### Graficando Distorciones

```python
fig, ax = plt.subplots(figsize=(5, 8))
ax.plot(df1["Drift"], df1["Story"], "bo-")
ax.plot(df1["Limit"], df1["Story"], "r--")
plt.title("Story Drifts", fontweight="bold")
plt.xlabel("Drift")
plt.ylabel("Stories")
ax.grid(lw=0.3)
plt.gca().invert_yaxis()
plt.xlim(0)
plt.show()
```
